```python
#Generation of marks
from random import randint
def generatemarks(n,lb,ub):
    filepath='DataFiles/marks.txt'
    with open(filepath,'w') as f:
        for i in range(n):
            r=randint(lb,ub)
            f.write(str(r)+'\n')
        print(n,"marks stored in file")
#generatemarks(30,1,100)


#class average
def classaverage(filepath):
    sum=0
    count=0
    with open(filepath,'r') as f:
        for i in f:
            sum+=int(i)
            count+=1
        avg=sum/count
    return avg

#classaverage('DataFiles/marks.txt')

# % of pass percentage
def passpercentage(filepath):
    count=0
    mc=0
    with open(filepath,'r') as f:
        for i in f:
            mc+=1
            if int(i)>=35:
                count+=1
    return ((count/mc)*100)
    
#passpercentage('DataFiles/marks.txt')


def failpercentage(filepath):
    count=0
    mc=0
    with open(filepath,'r') as f:
        for i in f:
            mc+=1
            if int(i)<35:
                count+=1
    return ((count/mc)*100)
    
#failpercentage('DataFiles/marks.txt')  


def distinction(filepath):
    count=0
    mc=0
    with open(filepath,'r') as f:
        for i in f:
            mc+=1
            if int(i)>=75:
                count+=1
    return ((count/mc)*100)
        
#distinction('DataFiles/marks.txt')

def highestfrequency(filepath):
    with open(filepath,'r') as f:
        sp=f.read().split()
        sp=list(map(int,sp))
        #sp=list(map(str,sp))
    return (sp.count(max(sp)))
    
#highestfrequency('DataFiles/marks.txt')

            
def lowestfrequency(filepath):
    with open(filepath,'r') as f:
        sp=f.read().split()
        sp=list(map(int,sp))
        #sp=list(map(str,sp))
    return (sp.count(min(sp)))
#lowestfrequency('DataFiles/marks.txt')   



def GenerateReport(filepath):
    while True:
        n=int(input("choose option:\n1)Generate marks\n2)Class Average\n3)passed percentage\n4)Failed percentage\n5)Distinction percentage\n6)Highest frequency\n7)Lowest frequency\n"))
        if n==1:
            st=int(input("Enter no of student marks"))
            generatemarks(st,1,100)
        elif n==2:
            print(classaverage(filepath))
        elif n==3:
            print(passpercentage(filepath))
        elif n==4:
            print(failpercentage(filepath))
        elif n==5:
            print(distinction(filepath))
        elif n==6:
            print(highestfrequency(filepath))
        elif n==7:
            print(lowestfrequency(filepath))
        else:
            break
    return

filepath='DataFiles/marks.txt'
GenerateReport(filepath)
```

```python
#Function to check if two strings are anagrams or not
# abc cba  ->True
# {a:1,b:1,c:1} {c:1,b:1,a:1}
# aabbcc ccbbaaa -->False
# {a:2,b:2,c:2} {a:3,b:2,c:2}

def CheckAnagram(s1,s2):
    if len(s1)!=len(s2):
        return False
    if sorted(s1)==sorted(s2):
        return True
    return False
#CheckAnagram("abc","cba")


def chardeletionsAnagrams(s1,s2):
    uncommon=[]     # To collect all uncommon characters-characters  only in a single string
    for i in s1:
        if i not in s2:
            uncommon.append(i)
    for i in s2:
        if i not in s1:
            uncommon.append(i)
    count=len(uncommon)
    freq1={}   #Frequency of common characters in string 1
    freq2={}
    uniqs1=[]  #unique characters in s1 and s2
    uniqs2=[]
    #Frequency of common unique characters in s1
    for i in s1:
        if i not in uncommon and i not in uniqs1:
            freq1[i]=s1.count(i)
            uniqs1.append(i)
    #Frequency of common unique characters in s1
    for i in s2:
        if i not in uncommon and i not in uniqs2:
            freq2[i]=s2.count(i)
            uniqs2.append(i)
    #Difference in frequencies for common characters and add that difference to the count
    for key in freq1.keys():
        count+=abs(freq1[key]-freq2[key])
    return count
t=int(input())
for i in range(t):
    s1=input()
    s2=input()
    print(chardeletionsAnagrams(s1,s2))
        
            
```
