```python
# Linear graph 
import matplotlib.pyplot as plt
plt.title("Basic linear graph")
plt.plot([1,2,3],[1,2,3])

plt.show
```

```python
# Graph to plot between distance and time

distance=list(range(2,11,2))
time=list(range(1,6))
# d=2*t
plt.title("Distance Vs Time")
plt.xlabel("Time")
plt.ylabel("distance")
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

```python
# v=at
t=list(range(1,11))
a=list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# Numbers vs cubes
n=list(range(10))
sqn=[i**3 for i in n]
plt.plot(n,sqn)
plt.grid()
plt.show()
```

```python
# voltage vs time
import numpy as np
#t=[i/10 for i in range(1,21)]
t=np.arange(0.1,2.0,0.1)
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title("Time vs voltage graph")

plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig("Images/Time-voltalge.png")
plt.show()
```

```python
# plot Income of Arkansas from 2005 to 2013
import pandas as pd
df=pd.read_csv("DataFiles/income.csv")
years=df.columns[2:]
incomeArkansas=df.values[3][2:]
plt.title("Income of Arkansas")
plt.xlabel('Year')
plt.ylabel("Income")
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Income of all states from 2005 to 2013 
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
    
plt.plot(years,incomeallstates)
plt.title("Income of all states")
plt.xlabel("year")
plt.ylabel("Income")
plt.show()
```
